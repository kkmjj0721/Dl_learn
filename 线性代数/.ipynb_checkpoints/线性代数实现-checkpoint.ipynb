线性代数：

标量由只有一个元素的张量表示

```python
import torch

x = torch.tensor([3.0])
y = torch.tensor([2.0])

x + y, x * y, x / y, x**y
```

可以将向量视为标量值组成的列表

```python
x = torch.arange(4)
x
```

可以通过张量的索引来访问任一元素

```python
x[3]
```

访问张量的长度

```python
len(x)
```

只有一个轴的张量，形状只有一个元素

```python
x.shape
```

通过指定两个分量m和n来创建一个形状为m✖n的矩阵

```python
A = torch.arange(20).reshape(5, 4)
A
```

矩阵的转置

```python
A.T
```

对称矩阵，A等于其转置 A = $A^T$

```python
B = torch.tensor([[1,2,3],[2,0,4],[3,4,5]])
B
```

```python
B == B.T
```

就像向量是标量的推广，矩阵是向量的推广一样，我们可以构建具有更多轴的数据结构

```python
X = torch.arange(24).reshape(2, 3, 4)                   # 2个大块，3行4列的小块
X
```

给定具有相同形状的任何两个张量，任何按元素二元运算的结果都将是相同形状的张量

```python
A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
B = A.clone()
A, A + B
```

两个矩阵的按元素乘法称为 哈达玛积，（数学符号为$\bigodot$）

```python
A * B
```

```python
a = 2
X = torch.arange(24).reshape(2, 3, 4)
a + X,(a * X).shape
```

计算其元素的和

```python
x = torch.arange(4.0)
x, x.sum()
```

表示任意形状张量的元素和

```python
A = torch.arange(20*2,dtype=torch.float32).reshape(2, 5, 4)
A,A.shape, A.sum()
```

指定求和汇总张量的轴

```python
A_sum_axis0 = A.sum(axis=0)                         # 沿着轴0求和，即将两个大块变成一个大块
A_sum_axis0, A_sum_axis0.shape
```

```python
A_sum_axis1 = A.sum(axis=1)                         # 沿着轴1求和，即将5行4列的小块变成1行4列的小块
A_sum_axis1, A_sum_axis1.shape
```

```python
A.sum(axis=[0, 1]), A.sum(axis=[0, 1]).shape           # 沿着轴0和轴1求和，即将两个大块和5行4列的小块都变成一个数
```

一个与求和相关的量是平均值（mean或average）

```python
A.mean(), A.sum() / A.numel()
```

```python
A.mean(axis=0), A.sum(axis=0) / A.shape[0]                 # 沿着轴0求平均值，即将两个大块变成一个大块
```

计算总和或均值时保持轴数不变

```python
sum_A = A.sum(axis=0, keepdim=True)                      # 沿着轴0求和，并保持轴数不变
sum_A, sum_A.shape
```

通过广播将 A 除以 sum_A 

```python
A / sum_A
```

某个轴计算 A 元素的累积求和

```python
A.cumsum(axis=0)                                       # 某个轴计算 A 元素的累积求和
```

点积是相同位置的按元素乘积的和

```python
y = torch.ones(4,dtype=torch.float32)
x,y, torch.dot(x,y)                                       # 点积是相同位置的按元素乘积的和
```

我们可以通过执行按元素乘法，然后进行求和来表示两个向量的点积

```python
torch.sum(x * y)
```

矩阵向量积 Ax

```python
A.shape,x.shape,torch.mv(A[0], x)                                       # 矩阵向量积 Ax
```

我们可以将矩阵-矩阵乘法AB看作是简单执行m次矩阵-向量积，并将结果拼接在一起

```python
B = torch.ones(4,3)
torch.mm(A[0],B)
```

L2范数是向量元素平方和的平方根：

$$||\mathbf{x}||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}$$

```python
u = torch.tensor([3.0,-4.0])
torch.norm(u)                                       # L2范数是向量元素平方和的平方根：
```

L1范数，它表示向量元素的绝对值之和

$$||\mathbf{x}||_1 = \sum_{i=1}^{n} |x_i|$$

```python
torch.abs(u).sum()                                       # L1范数，它表示向量元素的绝对值之和
```

矩阵的弗罗贝尼乌斯范数是矩阵元素的平方和的平方根：

$$||\mathbf{X}||_F = \sqrt{\sum_{i=1}^m \sum_{j=1}^n x_{ij}^2}$$

```python
torch.norm(torch.ones((4,9)))                               # 矩阵的Frobenius范数是矩阵元素平方和的平方根
```
